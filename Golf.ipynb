```python
eagle.drop(index=0,inplace=True)
birdie.drop(index=0,inplace=True)
bogey.drop(index=0,inplace=True)

bogey.rename(columns={0:'Name', 1:'Bogey', 2:'Holes'}, inplace=True)
birdie.rename(columns={0:'Name', 1:'Birdie', 2:'Holes'}, inplace=True)
eagle.rename(columns={0:'Name', 2:'Eagle', 1:'Holes'}, inplace=True)
```

```python
eagle.head()
```

```python
#Creating a combined data frame for all the shots

golf=pd.DataFrame()
Name=eagle['Name']
Eagle=eagle['Eagle'].astype(int)
Birdie=birdie['Birdie'].astype(int)
Bogey=bogey['Bogey'].astype(int)
Holes=eagle['Holes'].astype(int)

golf.insert(0, "Name", Name)
golf.insert(1, "Holes", Holes)
golf.insert(2, "Eagle", Eagle)
golf.insert(3, "Birdie", Birdie)
golf.insert(4, "Bogey", Bogey)
```

```python
golf.head()
```

```python
Par=Holes-(Eagle+Birdie+Bogey)
golf.insert(4, "Par", Par)
```

```python
golf.head()
```

```python
EagleRatio = Eagle/Holes
BirdieRatio = Birdie/Holes
ParRatio = Par/Holes
BogeyRatio = Bogey/Holes

golf.insert(6, "Eagle Ratio", EagleRatio)
golf.insert(7, "Birdie Ratio", BirdieRatio)
golf.insert(8, "Par Ratio", ParRatio)
golf.insert(9, "Bogey Ratio", BogeyRatio)
```

```python
golf.head()
```

```python
SkillRatio=EagleRatio*(0.6)+BirdieRatio*(0.4)

golf.insert(10, "Skill Ratio", SkillRatio)
```

```python
#Segregating Players based on their performance in the tournament

filter1=golf["Skill Ratio"] >= (golf_stats.loc['75%']['Skill Ratio'])
AdvancedPlayers=golf[filter1]
Advanced_Stats=AdvancedPlayers.describe()
```

```python
filter2=golf["Skill Ratio"] <= (golf_stats.loc['25%']['Skill Ratio'])
BeginnerPlayers=golf[filter2]
Beginner_Stats=BeginnerPlayers.describe()
```

```python
filter3=golf["Skill Ratio"] > (golf_stats.loc['25%']['Skill Ratio'])
filter4=golf["Skill Ratio"] < (golf_stats.loc['75%']['Skill Ratio'])
filter5=filter3*filter4
IntermediatePlayers=golf[filter5]
Intermediate_Stats=IntermediatePlayers.describe()
```

```python
#Lists regarding Probability of shots played by Advanced, Intermediate and Beginner players

p1=[Advanced_Stats.loc['mean']['Eagle Ratio'],Advanced_Stats.loc['mean']['Birdie Ratio'],Advanced_Stats.loc['mean']['Par Ratio'],Advanced_Stats.loc['mean']['Bogey Ratio']]
p2=[Intermediate_Stats.loc['mean']['Eagle Ratio'],Intermediate_Stats.loc['mean']['Birdie Ratio'],Intermediate_Stats.loc['mean']['Par Ratio'],Intermediate_Stats.loc['mean']['Bogey Ratio']]
p3=[Beginner_Stats.loc['mean']['Eagle Ratio'],Beginner_Stats.loc['mean']['Birdie Ratio'],Beginner_Stats.loc['mean']['Par Ratio'],Beginner_Stats.loc['mean']['Bogey Ratio']]
```

```python
#Let's begin playing Golf
#First we define various functions used during the game

#This function returns shots probabilistically based on the skills of the players

def shot(skill):
    shots=['Eagle','Birdie','Par','Bogey']
    if(skill=='A'):
        val=rand.choices(shots,p1)
    elif(skill=='I'):
        val=rand.choices(shots,p2)
    else:
        val=rand.choices(shots,p3)

    return val
```

```python
#This functions looks after how many points should be awarded for a particular shot
#Remember, lesser the points, more chance is to win :)

def point(shotplayed):
    if (shotplayed==['Eagle']):
        value=-2
    elif (shotplayed==['Birdie']):
        value=-1
    elif(shotplayed==['Par']):
        value=0
    else:
        value=1

    return value
```
